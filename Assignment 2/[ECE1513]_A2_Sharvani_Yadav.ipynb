### Part 1: Linear Regression with Scalar Inputs and Outputs

```python
# Part 1.1
# Plotting the dataset D to see the relationship between x and t

import numpy as np
import matplotlib.pyplot as plt

# Set up the dataset with x as inputs and t as outputs
x = np.array([1, 2, 3, 4, 5, 6, 7])       # x values
t = np.array([6, 4, 2, 1, 3, 6, 10])      # t values

# Print them to double-check the data
print("Inputs (x):", x)
print("Outputs (t):", t)

# Make a scatter plot using matplotlib
plt.figure(figsize=(5, 4))
plt.scatter(x, t, s=60, color='blue', edgecolor='k')  # plot each point
plt.title("Part 1.1 - Scatter Plot of Dataset D")
plt.xlabel("x (input)")
plt.ylabel("t (output)")
plt.grid(alpha=0.3)
plt.show()
```
